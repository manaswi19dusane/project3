```python

df['wip'] = df['wip'].fillna(df['wip'].median())
df 
```

```python

df.rename(columns={'department_finishing': 'department_finishing_1'}, inplace=True)
df
```

```python
df['productivity_ratio'] = df['actual_productivity'] / df['targeted_productivity']
df
```

```python
df.sort_values(by='productivity_ratio', ascending=False).head(10)
```

```python
df[['team','productivity_ratio']].groupby('team').mean().plot(kind='bar')
plt.ylabel("Avg Productivity Ratio") 
plt.show()
```

```python
df[df['over_time'] <= 2000]['actual_productivity'].mean()
df[df['over_time'] > 2000]['actual_productivity'].mean() 
df
```

```python
import matplotlib.pyplot as plt

# scatter plot with custom color (e.g., red)
plt.scatter(df['over_time'], df['actual_productivity'], color='yellow')

# change axis label text + color
plt.xlabel("Overtime (minutes)", color='blue')
plt.ylabel("Actual Productivity", color='green')

# change tick label colors
plt.xticks(color='purple')
plt.yticks(color='orange')

plt.show()
```

```python
team_perf = df.groupby('team')['actual_productivity'].mean()
dept_perf = df.groupby('department_sweing')['actual_productivity'].mean()
df
```

```python
team_perf.plot(kind='bar', title="Team-wise Average Productivity") 
plt.show()
```

```python
days = ['day_Monday','day_Tuesday','day_Wednesday','day_Thursday','day_Saturday','day_Sunday']
df_day = df[days + ['actual_productivity']].melt(id_vars='actual_productivity', var_name='Day', value_name='Flag')
df_day = df_day[df_day['Flag']==1]
df_day.groupby('Day')['actual_productivity'].mean().plot(kind='bar', title="Productivity by Day") 
plt.show()
```

```python
plt.scatter(df['over_time'], df['actual_productivity'])
plt.xlabel("Overtime") 
plt.ylabel("Actual Productivity")
plt.title("Overtime vs Productivity") 
plt.show()
```

```python
plt.scatter(df['incentive'], df['actual_productivity']) 
plt.xlabel("Incentive") 
plt.ylabel("Actual Productivity")
plt.title("Incentive vs Productivity") 
plt.show()
```
